```python
import os
import pandas as pd
import numpy as np
import geopandas as gpd
import folium

import time
import datetime as dt
from datetime import timedelta, datetime, timezone
from dateutil.relativedelta import relativedelta
import warnings
warnings.filterwarnings('ignore')

import plotly
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
data_folder_name = "data"

if not os.path.exists("VDS_results"):
    os.makedirs("VDS_results")
    os.makedirs("VDS_results/도솔대교_VDS_EtoW")
    os.makedirs("VDS_results/도솔대교_VDS_WtoE")

if not os.path.exists("data"):
    os.makedirs("data")
    os.makedirs("data/202302")
    os.makedirs("data/202303")
    os.makedirs("data/202304")
    os.makedirs("data/202402")
    os.makedirs("data/202403")
    os.makedirs("data/202404")
```

```python
# new 2023 files preprocessing

vds_link = pd.read_csv('[MERGED]total_VDS_with_osm.csv', encoding='cp949')

vds_sensors = ['VDS0022', 'VDS0023', 'VDS0103', 'VDS0104', 'VDS0107', 'VDS0108', 'VDS0109', 'VDS0110', 
               'VDS0112', 'VDS0113', 'VDS0123',  'VDS0124']

folder = ['202302', '202303', '202304']
filename = ['20230216', '20230316', '20230406']

data_to_use = ['vds_tr_vol_5mn', 'vds_trvl_spd_5mn', 'vds_trvl_tm_5mn']

for i in range(3):
    result = pd.DataFrame(columns = ['reg_ymdhms', 'vds_id', 'tr_vol', 'trvl_spd', 'trvl_tm'])
    
    vds_data = pd.read_csv(data_folder_name + '/' + folder[i] + '/' + filename[i] + '.csv')

    for sensor in vds_sensors:
        LINK_ID_da = vds_link.loc[vds_link['VDS_ID'] == sensor]['LINK_ID_da'].iloc[0]
        LINK_ID = vds_link.loc[vds_link['VDS_ID'] == sensor]['LINK_ID'].iloc[0]
    
        sensordata = vds_data.loc[(vds_data['link_id'] == LINK_ID) & (vds_data['road_route_id'])]
    
        for index, row in sensordata.iterrows():
            data = [{
                'reg_ymdhms' : str(row['ymd_id']) + str(row['hh_id']).zfill(2) + str(row['mn_id']).zfill(2) + '00',
                'vds_id' : sensor,
                'tr_vol' : row['vds_tr_vol_5mn'], 'trvl_spd' : row['vds_trvl_spd_5mn'],
                'trvl_tm' : row['vds_trvl_tm_5mn']
            }]
    
            resapp = pd.DataFrame(data)
            result = pd.concat([result, resapp], ignore_index=True)

    result.to_csv(data_folder_name + '/' + folder[i] + '/' + filename[i] + '_new.csv', index=False, encoding='utf-8-sig')
```

```python
vdsIDs = []
vds_coord = '도솔대교_VDS_좌표.csv'
df = pd.read_csv(vds_coord)

vds_id = '도솔대교_VDS_ID.csv'
ids = pd.read_csv(vds_id)

files = ['202302', '202303', '202304', '202402', '202403', '202404']
filenames = ['20230216_new', '20230316_new', '20230406_new', '20240215', '20240314', '20240404']

for index, row in df.iterrows():
    vdsIDs.append(row['vds_id'])
```

```python
peak = ['오전첨두', '비첨두', '오후첨두']
times = [['07:00~08:00', '08:00~09:00', '07:00~09:00'],
         ['12:00~13:00', '13:00~14:00', '12:00~14:00'],
         ['17:00~18:00', '18:00~19:00', '17:00~19:00']]

months = ['2월', '3월', '4월']

EtoW = ['VDS0022', 'VDS0103', 'VDS0109', 'VDS0123']
WtoE = ['VDS0023', 'VDS0104', 'VDS0110', 'VDS0124']

directionstr = ['EtoW', 'WtoE']
directionrev = [WtoE, EtoW]
directionfolder = ['도솔대교_VDS_EtoW', '도솔대교_VDS_WtoE']
```

```python
d = 0
for direc in directionstr:
    print(direc)
    for i in range(3):
        #month
        print(filenames[i] + ' and ' + filenames[i+3])
        data_2023 = pd.read_csv('data' + '/' + files[i] + '/' + filenames[i] + '.csv')
        data_2024 = pd.read_csv('data' + '/' + files[i+3] + '/' + filenames[i+3] + '.csv')
    
        date2023 = filenames[i]
        date2024 = filenames[i+3]
    
        data_2023['reg_ymdhms'] = pd.to_datetime(data_2023['reg_ymdhms'], format ='%Y%m%d%H%M%S')
        data_2024['reg_ymdhms'] = pd.to_datetime(data_2024['reg_ymdhms'], format ='%Y%m%d%H%M%S')
    
        for j in range(3):
            #peak
            print(peak[j])
            for k in range(3):
                #times
                #print(times[j][k])

                result = pd.DataFrame(columns = ['route_id', 'VDS_id', 'longitude', 'latitude', 'movers_2023', 'movers_2024', 
                                                 'avg_speed_2023', 'avg_speed_2024',
                                                 'movers_diff', 'avg_speed_diff', 'location'])
                
                route_id = 0
                for vds in vdsIDs:
                    if (vds not in directionrev[d]):
                        #print(vds)
                        vds_row = ids.loc[(ids['VDS 아이디'] == vds)]
                        section = vds_row['VDS 구간 아이디'].iloc[0]
                        detr = vds_row['검지기 아이디'].to_list()

                        #data_rows_2023 = data_2023.loc[(data_2023['vds_id'] == vds) & (data_2023['detr_id'].isin(detr)) & (data_2023['reg_ymdhms'].dt.time.between(pd.Timestamp(times[j][k].split('~')[0]).time(), pd.Timestamp(times[j][k].split('~')[1]).time()))]
                        data_rows_2023 = data_2023.loc[(data_2023['vds_id'] == vds) & (data_2023['reg_ymdhms'].dt.time.between(pd.Timestamp(times[j][k].split('~')[0]).time(), pd.Timestamp(times[j][k].split('~')[1]).time()))]
                        data_rows_2024 = data_2024.loc[(data_2024['vds_id'] == vds) & (data_2024['detr_id'].isin(detr)) & (data_2024['reg_ymdhms'].dt.time.between(pd.Timestamp(times[j][k].split('~')[0]).time(), pd.Timestamp(times[j][k].split('~')[1]).time()))]

                        movers_2023 = data_rows_2023['tr_vol'].sum()
                        if(movers_2023 == 0):
                            print(vds + ' 2023 ' + months[i] + '_' + peak[j] + '_' + (times[j][k]).replace(':00', '') + ' no data')
                        movers_2024 = data_rows_2024['tr_vol'].sum()
                        if(movers_2024 == 0):
                            print(vds + ' 2024 ' + months[i] + '_' + peak[j] + '_' + (times[j][k]).replace(':00', '') + ' no data')

                        movers_diff = movers_2024 - movers_2023

                        avg_speed_2023 = (data_rows_2023.loc[(data_rows_2023['trvl_spd'] != 0)])['trvl_spd'].mean()
                        if(np.isnan(avg_speed_2023) == True):
                            avg_speed_2023 = 0
                        avg_speed_2024 = (data_rows_2024.loc[(data_rows_2024['trvl_spd'] != 0)])['trvl_spd'].mean()
                        if(np.isnan(avg_speed_2024) == True):
                            avg_speed_2024 = 0
                        avg_speed_diff = avg_speed_2024 - avg_speed_2023


                        long = df.loc[(df['vds_id'] == vds)]['posx'].iloc[0]
                        lat = df.loc[df['vds_id'] == vds]['posy'].iloc[0]
                        
                        data = [{'route_id' : route_id, 'VDS_id' : vds, 'longitude' : long, 'latitude' : lat,
                                 'movers_2023' : movers_2023, 'movers_2024' : movers_2024, 'avg_speed_2023' : avg_speed_2023, 
                                 'avg_speed_2024' : avg_speed_2024, 'movers_diff' : movers_diff, 'avg_speed_diff' : avg_speed_diff,
                                 'location' : df.loc[df['vds_id'] == vds]['inst_locat'].iloc[0]
                                }]

                        resapp = pd.DataFrame(data)
                        result = pd.concat([result, resapp], ignore_index=True)
                        
                        route_id += 1

                filepeaktime = (times[j][k]).replace(':00', '')
                filename = 'VDS' + '_' + months[i] + '_' + peak[j] + '_' + filepeaktime + '_' + directionstr[d] + '.csv'

                result_name = 'VDS_results/' + directionfolder[d] + '/' + filename
                result.to_csv(result_name, index=False, encoding='utf-8-sig')
    d += 1

print('done')
```
